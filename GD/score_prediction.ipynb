```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the dataset
x = np.array([1, 2, 3])  # Hours Studied (minutes)
y = np.array([2, 3, 5]) #  Test Score
```

```python
# Initialize parameters
theta_0 = 0  # Intercept
theta_1 = 0  # Slope
learning_rate = 0.1  # Learning rate
iterations = 5  # Number of iterations
m = len(x)  # Number of data points
```

```python
cost_history = []  # Store cost values
```

```python
# Perform Gradient Descent
for i in range(iterations):
    # Compute predictions
    y_pred = theta_0 + theta_1 * x
    
    # Compute errors
    errors = y_pred - y
    
    # Compute cost (Mean Squared Error)
    cost = (1/(2*m)) * np.sum(errors**2)
    cost_history.append(cost)  # Store cost value
    
    # Compute gradients
    grad_theta_0 = (1/m) * np.sum(errors)
    grad_theta_1 = (1/m) * np.sum(errors * x)
    
    # Update parameters
    theta_0 = theta_0 - learning_rate * grad_theta_0
    theta_1 = theta_1 - learning_rate * grad_theta_1

    # print("Iteration: " + str(i) + " theta0: " + str(theta_0) + "theta1: "+ str(theta_1))
    print(f"Iteration {i+1}: theta_0 = {theta_0:.4f}, theta_1 = {theta_1:.4f}, Cost = {cost:.4f}")
```

```python
# Plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Training Data')
plt.plot(x, theta_0 + theta_1 * x, color='red', label='Regression Line')
plt.xlabel("Hours Studied")
plt.ylabel("Test Score")
plt.title("Linear Regression Fit")
plt.legend()
plt.show()
# Plot Cost Function Convergence
plt.figure(figsize=(8, 6))
plt.plot(range(1, iterations+1), cost_history, marker='o', linestyle='-')
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Cost Function Convergence")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function and its derivative
def f(x):
    return x**3  # Our function: f(x) = x^3
```

```python
# Define its derivative
def df(x):
    return 3*x**2  # Derivative: f'(x) = 3x^2
```

```python
# Gradient Descent Parameters
learning_rate = 0.1  # Step size
x = np.random.uniform(-5, 5)  # Start at a random point
# iterations = 20  # Number of steps
```

```python
tolerance = 1e-6  # Stop when gradient is very small
```

```python
# Store values for visualization
x_values = [x]
y_values = [f(x)]
```

```python
# # Perform Gradient Descent with stops 
# for _ in range(iterations):
#     gradient = df(x)
#     x = x - learning_rate * gradient  # Update step
#     x_values.append(x)
#     y_values.append(f(x))
```

```python
# Perform Gradient Descent with tolerance 
while abs(df(x)) > tolerance:
    gradient = df(x)
    x = x - learning_rate * gradient  # Update step
    x_values.append(x)
    y_values.append(f(x))
```

```python
# Plot the function
x_range = np.linspace(-5, 5, 100)
y_range = f(x_range)
plt.plot(x_range, y_range, label='f(x) = x^3')
plt.scatter(x_values, y_values, color='red', label='Gradient Descent Steps')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Gradient Descent Visualization')
plt.show()
```

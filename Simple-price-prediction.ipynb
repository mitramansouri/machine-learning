```python
import numpy as np
import matplotlib.pyplot as plt

# Define the dataset
x = np.array([5, 10, 15])  # Exercise Time (minutes)
y = np.array([10, 18, 26])  # Calories Burned

# Initialize parameters
theta_0 = 0  # Intercept
theta_1 = 0  # Slope
learning_rate = 0.01  # Learning rate
iterations = 10  # Number of iterations
m = len(x)  # Number of data points
cost_history = []  # Store cost values for each iteration

# Plot setup
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Training Data')  # Plot training data

# Perform Gradient Descent
for i in range(iterations):
    # Compute predictions
    y_pred = theta_0 + theta_1 * x
    
    # Compute errors
    errors = y_pred - y
    
    # Compute cost (Mean Squared Error)
    cost = (1/(2*m)) * np.sum(errors**2)
    cost_history.append(cost)  # Store cost value
    
    # Compute gradients
    grad_theta_0 = (1/m) * np.sum(errors)
    grad_theta_1 = (1/m) * np.sum(errors * x)
    
    # Update parameters
    theta_0 = theta_0 - learning_rate * grad_theta_0
    theta_1 = theta_1 - learning_rate * grad_theta_1
    
    # Plot regression lines for each iteration
    label = f'Iteration {i+1}' if i in [0, iterations//2, iterations-1] else "_nolegend_"
    plt.plot(x, theta_0 + theta_1 * x, linestyle='--', alpha=0.5, label=label)

# Final plot adjustments
plt.xlabel("Exercise Time (minutes)")
plt.ylabel("Calories Burned")
plt.title("Gradient Descent Progress")
plt.legend()  # Ensure only labeled lines appear in the legend
plt.show()
```
